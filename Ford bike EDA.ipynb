## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
import pandas as pd

# Load data
df = pd.read_csv('/content/201801-fordgobike-tripdata.csv.zip')

# --------------------
# 1. Convert date columns to datetime
# --------------------
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])

# --------------------
# 2. Handle missing values
# --------------------
# Fill missing gender with 'Unknown'
df['member_gender'] = df['member_gender'].fillna('Unknown')

# Fill missing birth year with median (or you can drop rows if preferred)
df['member_birth_year'] = df['member_birth_year'].fillna(df['member_birth_year'].median())

# --------------------
# 3. Feature Engineering
# --------------------
# Calculate trip duration in minutes
df['duration_min'] = df['duration_sec'] / 60

# Calculate age of the member
df['member_age'] = 2018 - df['member_birth_year']  # Based on dataset year

# --------------------
# 4. Rename columns (optional)
# --------------------
df.rename(columns={
    'start_time': 'trip_start_time',
    'end_time': 'trip_end_time',
    'user_type': 'user_category'
}, inplace=True)

# --------------------
# 5. Drop duplicates (if any)
# --------------------
df.drop_duplicates(inplace=True)

# --------------------
# 6. Final check
# --------------------
print(df.info())
print(df.head())
````
